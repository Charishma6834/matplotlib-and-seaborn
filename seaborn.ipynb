# DATA VISUALISATION WITH MATPLOTLIB AND SEABORN USING THE IRIS DATASET

**Matplotlib :**

Matplotlib is a powerful and widely-used Python library for creating static, animated, and interactive visualizations. It is highly flexible and allows users to generate a wide range of plots, including line graphs, bar charts, scatter plots, histograms, pie charts, and more.

 __Uses of matplotlib :__
 - *Quick and Easy Plotting* : You can create basic plots in just a few lines of code.
 - *Customization* : Matplotlib allows for highly customizable plots with various styles, themes, and colors.
 - *Extensive Documentation* : Matplotlib has extensive documentation and community support, making it easy to learn and use.
 - *Compatibility* : It works well in a wide variety of environments, including Jupyter Notebooks, which is great for data exploration.

 Matplotlib is a foundational tool for data visualization in Python, offering both simplicity for basic plots and powerful control for more sophisticated visualizations.

**Seaborn :**

  Seaborn is a powerful and easy-to-use Python library for data visualization built on top of Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. While Matplotlib focuses on providing fine-grained control over plots, Seaborn simplifies the process by offering built-in themes and color palettes for creating visually appealing plots with minimal effort.

**Uses of seaborn** :
 - *Simplifies Complex Plots* : Seaborn makes it easy to create complex visualizations with very little code.
 - *Built-in Themes and Color Palettes* : It automatically applies attractive styles and color schemes.
 - *Statistical Inference* : Seaborn can handle many statistical plots with built-in aggregations.
 - *Integrates with Matplotlib* : Seaborn plots are Matplotlib objects, so you can further customize them using Matplotlib functions.

Seaborn is a high-level library that helps you create beautiful, easy-to-read statistical plots with less effort than Matplotlib, while still allowing for advanced customization when needed. Itâ€™s an ideal choice when working with datasets and performing exploratory data analysis (EDA).

**IRIS DATASET :**
The Iris dataset is one of the most famous datasets used for machine learning and data analysis.

It consists of 150 observations of iris flowers from three species: Setosa, Versicolor, and Virginica.

Each observation includes four features:
- Sepal length
- Sepal width
- Petal length
- Petal width
  
These features are numeric, and the goal is often to classify the flower species based on these measurements. The dataset is popular due to its simplicity and usefulness in visualizing relationships between features.

GENERAL STATISTICS PLOT :
 -

Here's a Python program that uses the Iris dataset to create a general statistical summary and a visualization using Seaborn's pairplot. 
- The pairplot helps visualize relationships between features, while pandas.describe() provides a numerical summary of the dataset.

**IMPORTING NECESSARY LIBRARIES :**


```python
!pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

**LOAD THE IRIS DATASET**:

```python
iris = sns.load_dataset("iris")
```

**STATISTICAL SUMMARY :**

```python
# statistical summary using pandas' describe()
print("Statistical Summary:\n", iris.describe())
```

**PAIRPLOT TO VISUALIZE RELATIONSHIPS BETWEEN VARIABLES AND SPECIES :**

```python
# pairplot to visualize relationships between variables and species
sns.pairplot(iris, hue="species", diag_kind="kde", markers=["o", "s", "D"],palette='Set2')
# Show the plot
plt.show()
```

PIE PLOT FOR SPECIES FREQUENCY :
 -

```python
# Get the frequency of each species
species_count = iris['species'].value_counts()

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(species_count, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'pink', '#d8bfd8'])
plt.title('FREQUENCY OF IRIS SPECIES',fontsize=16)
plt.legend(title='SPECIES',loc='lower right',labels=species_count.index)
# Display the pie chart
plt.show()
```

RELATIONSHIP BETWEEN SEPAL LENGTH AND WIDTH :
-

```python
#scatter plot for Sepal Length vs Sepal Width
plt.figure(figsize=(8, 6))
plt.scatter(iris['sepal_length'], iris['sepal_width'], c='purple', alpha=0.7)

# Add labels and title
plt.title('Relationship between Sepal Length and Sepal Width',fontsize=16)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Display the scatter plot
plt.show()
```

DISTRIBUTION OF SEPAL AND PETAL FEATURES :
-

```python
# Set the figure size for the plot
plt.figure(figsize=(10, 8))

# Create subplots for each feature's distribution
plt.subplot(2, 2, 1)
plt.hist(iris['sepal_length'], bins=20, color='#d8bfd8', alpha=0.7,edgecolor='black')
plt.title('Sepal Length Distribution')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')

plt.subplot(2, 2, 2)
plt.hist(iris['sepal_width'], bins=20, color='pink', alpha=0.7,edgecolor='black')
plt.title('Sepal Width Distribution')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')

plt.subplot(2, 2, 3)
plt.hist(iris['petal_length'], bins=20, color='coral', alpha=0.7,edgecolor='black')
plt.title('Petal Length Distribution')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')

plt.subplot(2, 2, 4)
plt.hist(iris['petal_width'], bins=20, color='lightblue', alpha=0.7,edgecolor='black')
plt.title('Petal Width Distribution')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plot
plt.show()
```

JOINTPLOT FOR SEPAL LENGTH VS SEPAL WIDTH :
-

```python
# Create a joint plot for Sepal Length vs Sepal Width
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='scatter', color='#dda0dd')

# Show the plot
plt.show()
```

KDE PLOT FOR SETOSA SEPICIES (Sepal Length vs Sepal Width):
-

```python
# Filter the dataset for Setosa species
setosa = iris[iris['species'] == 'setosa']

# Create a KDE plot for Sepal Length vs Sepal Width
plt.figure(figsize=(8, 6))
sns.kdeplot(x='sepal_length', y='sepal_width', data=setosa, cmap='Blues', fill=True)

# Add labels and title
plt.title('KDE Plot of Sepal Length vs Sepal Width for Setosa',fontsize=15)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Show the plot
plt.show()
```

KDE PLOT FOR SETOSA SPECIES (Petal Length vs Petal Width) :
-

```python

# Filter the dataset for Setosa species
setosa = iris[iris['species'] == 'setosa']

# Create a KDE plot for Petal Length vs Petal Width
plt.figure(figsize=(8, 6))
sns.kdeplot(x='petal_length', y='petal_width', data=setosa, cmap='Blues', fill=True)

# Add labels and title
plt.title('KDE Plot of Petal Length vs Petal Width for Setosa',fontsize=15)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')

# Show the plot
plt.show()
```
